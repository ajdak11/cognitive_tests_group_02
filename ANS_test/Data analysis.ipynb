```python
from scipy import stats
from scipy.stats import sem
import numpy as np 

#Reading the csv file
ANS_df = pd.read_csv('BIOS0030 group project results  (Responses) - Form responses 1.csv')

#Creating new dataframes for the different genders 
male_df = ANS_df[ANS_df.gender == 'm']
female_df = ANS_df[ANS_df.gender == 'f'] 
others_df = ANS_df[ANS_df.gender == 'o'] 

#calculating mean for total scores of each gender 
m_mean = male_df['score'].mean()
f_mean = female_df['score'].mean()

#Shapiro-wilk test for normality of data 
m_shapiro = stats.shapiro(male_df['score'])
f_shapiro = stats.shapiro(female_df['score'])

#Calculating variances of total scores of each gender 
m_var = np.var(male_df['score'])
f_var = np.var(female_df['score'])

#Calculating t test 
t_test = stats.ttest_ind(male_df['score'], female_df['score'], equal_var = False) 

#Calculating standard error of mean 
m_SE = sem(male_df['score']) 
f_SE = sem(female_df['score']) 

data_table = {
    'Gender' : ['male','female'], 
    'mean' : [m_mean, f_mean], 
    'SEM' : [m_SE,f_SE], 
    'Score difference' : [m_mean-f_mean, f_mean-m_mean],
    't test statistic, p-value' : [t_test, t_test], 
} 

df = pd.DataFrame(data=data_table)
df



```
